## Gr√°fico de Resultados en Kaggle

```python
import pandas as pd

# Datos para el DataFrame
data = {
    'Public Score': [0.519, 0.609, 0.6, 0.605, 0.63, 0.635, 0.6, 0.596, 0.6, 0.639, 0.639, 0.639, 0.6, 0.618, 0.66],
    'Private Score': [0.544, 0.589, 0.551, 0.602, 0.57, 0.583, 0.551, 0.57, 0.525, 0.57, 0.57, 0.57, 0.557, 0.583, 0.602]
}

# Crea el DataFrame
df = pd.DataFrame(data)

# Muestra el DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# Crea el gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(df['Public Score'], label='Public Score')
plt.plot(df['Private Score'], label='Private Score')
plt.title('Evolution of Public and Private Scores')
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.xticks(range(0, len(df), 1))  # Ajusta los ticks del eje x para que vayan de 1 en 1
plt.legend()

# Muestra el gr√°fico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crea el gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(df['Public Score'], label='Public Score')
plt.plot(df['Private Score'], label='Private Score')

# A√±ade las puntuaciones a cada punto
for i, score in enumerate(df['Public Score']):
    plt.text(i, score, str(score), ha='center', va='bottom')
for i, score in enumerate(df['Private Score']):
    plt.text(i, score, str(score), ha='center', va='top')

plt.title('Evolution of Public and Private Scores')
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.xticks(range(0, len(df), 1))  # Ajusta los ticks del eje x para que vayan de 1 en 1
plt.legend()

# Muestra el gr√°fico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calcula las medias
public_score_mean = df['Public Score'].mean()
private_score_mean = df['Private Score'].mean()

# Crea el gr√°fico de barras
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Public Score'], label='Public Score')
plt.bar(df.index, df['Private Score'], label='Private Score', alpha=0.5)

# A√±ade las l√≠neas de la media
plt.axhline(public_score_mean, color='blue', linestyle='dashed', linewidth=1)
plt.axhline(private_score_mean, color='orange', linestyle='dashed', linewidth=1)

# A√±ade las etiquetas de la media
plt.text(len(df), public_score_mean, f'Public Score Mean: {public_score_mean:.3f}', va='center', ha='right', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
plt.text(len(df), private_score_mean, f'Private Score Mean: {private_score_mean:.3f}', va='center', ha='right', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.title('Public and Private Scores with Mean')
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.xticks(np.arange(0, len(df), step=1))  # Ajusta los ticks del eje x para que vayan de 1 en 1
plt.legend()

# Muestra el gr√°fico
plt.show()
```
